## 7. Matplotlib

#### 7(a)

```python
# create two one dimensional arrays x and y and plot y vs x, add title, xlabel, ylabel, grid
x = np.linspace(-5, 5, num=20)
y = np.array([j ** 2 for j in x])
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph')
plt.grid(ls='-.')
plt.show()
```

#### 7(b)

```python
# create multiple arrays and plot them with different styles, add legends, add text/mathematical equations on the plot
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])

plt.plot(x, y1, 'ys-')
plt.text(-1.5, -2, r'equation: $Y = X$', fontsize = 10)
plt.xlabel('x')
plt.ylabel('y1')
plt.show()

plt.plot(x, y2, 'go--')
plt.text(-1.5, 5, r'equation: $Y = X^2$', fontsize = 10)
plt.xlabel('x')
plt.ylabel('y2')
plt.show()

plt.plot(x, y3,'bH-.')
plt.text(-1.5, -20, r'equation: $Y = X^3$', fontsize = 10)
plt.xlabel('x')
plt.ylabel('y3')
plt.show()
```

#### 7(c)

```python
# create multiple subplots, play around with the figure size, text font/size
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])

fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(7, 14))
fig.suptitle('Subplots')
ax1.plot(x, y1)
ax1.set_title('Y=X')

ax2.plot(x, y2)
ax2.set_title('Y=X^2')

ax3.plot(x, y3)
ax3.set_title('Y=X^3')

plt.show()
```

#### 7(d)

```python
# get familiar with get current axis (gca) handle to do the above tasks
plt.gca().plot([0,1,4])
```

#### 7(e)

```python
# change the limits on x and y axes, use logarithmic axes to plot
x = np.linspace(-5, 5, num=20)
y = np.array([j ** 2 for j in x])

fig, ax = plt.subplots()

ax.plot(x, y)

ax.set_xscale('log')
ax.set_yscale('log')

plt.xlabel('logx')
plt.ylabel('logy')

plt.show()
```

## 8. Seaborn

#### 8(a)-8(b)

```python
# use the Salaries.csv file in Pandas section
data = pd.read_csv('Salaries.csv', index_col='playerID')

# create a dataframe
df=pd.DataFrame(data)
df.columns = ['yearID', 'teamID', 'lgID', 'salary']
df.head()
```

```python
# and try to plot it with seaborn
sns.scatterplot(x="yearID", y="salary", data=df)
```

#### 8(c)

```python
# perform statistical estimation on the data using seaborn in-built functions
# lmplot
g = sns.lmplot(x="yearID", y="salary", data=df)
```

```python
# catplot
g = sns.catplot(x="teamID", y="salary", col='lgID', data=df)
```

```python
# relplot
g = sns.relplot(x="yearID", y="salary", hue='teamID', col='lgID', data=df)
```

#### 8(d) create axis level functions like boxplot to visualize

```python
# create axis level functions like boxplot to visualize
df1 = df
df1['salary'] = df['salary'].astype(int)

fig, ax = plt.subplots(figsize=(7,5))
sns.boxplot(x='teamID', y='salary', data=df)
fig.subplots_adjust(top=0.9)
```

#### 8(e) visualize the dataset structure using pairplot and jointplot

```python
# visualize the dataset structure using pairplot and jointplot
# jointplot
sns.jointplot(data=df, x="yearID", y="salary")
```

```python
# pairplot
var_set = [
    "yearID",
    "teamID",
    "lgID",
    "playerID",
    "salary"
]

head_set = []
head_set.extend(var_set)
head_set.append("num_feat")

df = pd.read_csv('Salaries.csv', index_col='playerID', header=None, names=head_set)

df['num_feat'] = 100 * np.random.random_sample(df.shape[0])

df_copy = df
cols_with_team_ATL = df_copy.loc[df_copy.teamID=="ATL", ]

# Create the default pairplot
pairplot_fig = sns.pairplot(cols_with_team_ATL, vars=['yearID', 'salary', 'num_feat'])
plt.subplots_adjust(top=0.9)

plt.show()
```
